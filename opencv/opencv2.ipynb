# 영상을 명암 영상으로 변환하고 반으로 축소하기

```python


import cv2 as cv
import sys

img = cv.imread('./opencv_data/soccer.jpg')
if img is None:                                 # img를 불러오지 못한 경우
    sys.exit('파일을 읽어 올 수 없습니다.')       # 프로그램 종료
# 칼라 이미지정보를 가지고 있는 img
img[0][0] # RGB 정보를 가지고 있는 배열의 1요소

# 색상 변환 (흑백-Gray로 변환)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)      # RGB 칼라를 GRAY로 변환하는 옵션
gray_small = cv.resize(gray, dsize=(0,0), fx=0.5, fy=0.5) # 사이즈를 반으로 축소

# 영상 정보를 파일로 저장
cv.imwrite('./opencv_data/soccer_gray.jpg', gray)
cv.imwrite('./opencv_data/soccer_gray_small.jpg', gray_small)

# 영상 이미지 보여주기 imshow
cv.imshow('Color image', img)
cv.imshow('Gray image', gray)
cv.imshow('Color image small', gray_small)

cv.waitKey()
cv.destroyAllWindows()
```

# 마우스를 통한 상호작용 

- 마우스로 클릭한 곳에 직사각형 그리기 (Callback)


```python

import cv2 as cv
import sys

img = cv.imread('./opencv_data/girl_laughing.jpg')
if img is None:                                 # img를 불러오지 못한 경우
    sys.exit('파일을 읽어 올 수 없습니다.')       # 프로그램 종료

#-----------------이미지 데이터 로드 ------------------------------------

# 함수구현 - callback함수 : 마우스 우클릭, 좌클릭시 사각형 표시하기
def draw(event, x, y, flags, param):
    if event == cv.EVENT_LBUTTONDOWN:   # 마우스 왼쪽 클릭시 (눌렀다뗏을때가 아니라 눌렀을때)
        cv.rectangle(img, (x,y),(x+200, y+200), (0,0,255),2) # 200x200 사이즈 레드 사각형
    elif event == cv.EVENT_RBUTTONDOWN: # 마우스 오른쪽 클릭시
        cv.rectangle(img,(x,y),(x+100, y+100), (255,0,0),2)  # 100x100 사이즈 블루 사각형
    
    cv.imshow('Drawing', img)
    
cv.namedWindow('Drawing')
cv.imshow('Drawing',img)

# Drawing 윈도우에 draw 콜백 함수 지정
cv.setMouseCallback('Drawing', draw)    # 콜백함수 설정

while(True):            # 마우스 이벤트가 언제 발생할지 모르기에 무한 반복해서 q를 입력시 닫히도록
    if cv.waitKey(1) == ord('q'):
        cv.destroyAllWindows()
        break
```

# 마우스 드래그로 도형 크기 조절하기

```python

import cv2 as cv
import sys

img = cv.imread('./opencv_data/girl_laughing.jpg')
if img is None:                                 # img를 불러오지 못한 경우
    sys.exit('파일을 읽어 올 수 없습니다.')       # 프로그램 종료

#-----------------이미지 데이터 로드 ------------------------------------

# 함수구현 - callback함수 : 마우스 드래그를 통해서 사각형 크기 지정하기
    # 마우스 드래그 - EVENT_LBUTTONDOWN, EVENT_LBUTTONUP
def draw(event, x, y, flags, param):
    global ix,iy    # 전역 변수 설정
    
    if event == cv.EVENT_LBUTTONDOWN:   # 마우스 왼쪽 버튼 클릭했을 때 초기 위치 저장
        ix, iy = x, y                   # 왼쪽 마우스 클릭시 값저장
    elif event == cv.EVENT_LBUTTONUP:   # 마우스 왼쪽 버튼 클릭했을 때 직사각형 그리기
        cv.rectangle(img, (ix,iy), (x, y), (0,0,255), 2)
        
    cv.imshow('Drawing', img)
    
cv.namedWindow('Drawing')
cv.imshow('Drawing',img)

# Drawing 윈도우에 draw 콜백 함수 지정
cv.setMouseCallback('Drawing', draw)    # 콜백함수 설정

while(True):            # 마우스 이벤트가 언제 발생할지 모르기에 무한 반복해서 q를 입력시 닫히도록
    if cv.waitKey(1) == ord('q'):
        cv.destroyAllWindows()
        break
```

# 페인팅 기능 만들기
- 빨간색 붓과 파란색 붓으로 페인팅하기

```python

import cv2 as cv
import sys

img = cv.imread('./opencv_data/soccer.jpg')
if img is None:                                 # img를 불러오지 못한 경우
    sys.exit('파일을 읽어 올 수 없습니다.')       # 프로그램 종료

# BrushSize
BrushSize = 5                                   # 붓의 크기
LColor,RColor=(255,0,0),(0,0,255)               # 파란색과 빨간색
#-----------------이미지 데이터 로드 ------------------------------------

# 함수구현 - 페인팅 callback함수 : 마우스 드래그를 통해서 페인팅하기
    # 마우스 드래그 - EVENT_LBUTTONDOWN, EVENT_LBUTTONUP
def painting(event, x, y, flags, param):
    print(event, x, y, flags, param)
    if event == cv.EVENT_LBUTTONDOWN:
        cv.circle(img,(x,y),BrushSize, LColor, -1)  # 마우스 왼쪽 버튼을 클릭했을 때 파란원을 그림
    elif event == cv.EVENT_RBUTTONDOWN:
        cv.circle(img, (x,y), BrushSize, RColor, -1) # 마우스 오른쪽 버튼을 클릭했을 때 빨간원을 그림
    elif event == cv.EVENT_MOUSEMOVE and flags == cv.EVENT_FLAG_LBUTTON:
        cv.circle(img,(x,y), BrushSize, LColor, -1)  # 마우스 왼쪽 버튼 클릭하고 이동하면서 파란색 원이 그려짐
    elif event == cv.EVENT_MOUSEMOVE and flags == cv.EVENT_FLAG_RBUTTON:
        cv.circle(img, (x,y), BrushSize, RColor, -1) # 마우스 오른쪽 버튼 클릭하고 이동하면서 파란색 원이 그려짐
    
    cv.imshow('Painting', img)
    
cv.namedWindow('Painting')
cv.imshow('Painting',img)

# Painting 윈도우에 draw 콜백 함수 지정
cv.setMouseCallback('Painting', painting)    # 콜백함수 설정

while(True):            # 마우스 이벤트가 언제 발생할지 모르기에 무한 반복해서 q를 입력시 닫히도록
    if cv.waitKey(1) == ord('q'):
        cv.destroyAllWindows()
        break
```

```python
import cv2 as cv
import numpy as np    #이미지는 np배열의 형태로 처리됨
import sys

connstr = 'libcamerasrc ! video/x-raw, width=640, height=480, framerate=30/1 ! appsink'
cap =  cv.VideoCapture(connstr, cv.CAP_GSTREAMER)  # 카메라 연결

#카메라 연결 검증
if cap.isOpened() == False:
  sys.exit('카메라 연결 실패')

# 캡쳐한 이미지 저장
frames = []
while True:
  ret, frame = cap.read()        # 비디오를 구성하는 프레임을 획득

  if not ret:
     print('프레임 획득에 실패하여 루프를 나갑니다.')

  cv.imshow('Video display', frame)

  # 이미지 캡쳐
  key = cv.waitKey(1)    # 1밀리초 동안 키보드 입력 기다림
  if key == ord('c'):
     frames.append(frame)
  elif key == ord('q'):  # 'q' 키가 들어오면 루프 나가기
     break

cap.release()            # 카메라와 연결 끊기
cv.destroyAllWindows()   # Window 끄기

# 캡쳐한 프레임을 병합
if len(frames) > 0:
  imgs = frames[0]
  for i in  range(1, min(3,len(frames))):  # 캡처 이미지 최대 3개까지 이어붙임
    imgs = np.hstack((imgs, frames[i]))
  cv.imshow('collected images', imgs)
  cv.waitKey()
  cv.destroyAllWindows()
```
